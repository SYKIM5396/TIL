```python
# 결측치를 0로 대체
df = df.fillna(0)
```

```python
#Month 컬럼을 달과 년도로 나눠줌
df["Month"], df["Year"] =df.Month.str.split(' ').str
```

```python
#컬럼 확인
df.columns
```

```python
#컬럼 위치 재정렬 
df= df[['Month', 'Year', 'Coconut Oil Price', 'Coconut Oil Price Change',
       'Olive Oil Price', 'Olive Oil Price Change', 'Palm Kernel Oil Price',
       'Palm Kernel Oil Price Change', 'Palm Oil Price',
       'Palm Oil Price Change', 'Peanut Oil Price', 'Peanut Oil Price Change',
       'Rapeseed Oil Price', 'Rapeseed Oil Price Change', 'Soybean Oil Price',
       'Soybean Oil Price Change', 'Sunflower Oil Price',
       'Sunflower Oil Price Change']]
df.head()
```

```python
# Month 컬럼의 달을 숫자로 바꿔주기
df.Month = df.Month.replace(
{
            'Jan' : 1,
            'Feb' : 2,
            'Mar' : 3,
            'Apr' : 4,
            'May' : 5,
            'Jun' : 6,
            'Jul' : 7,
            'Aug' : 8,
            'Sep' : 9,
            'Oct' : 10,
            'Nov' : 11,
            'Dec' : 12
    } )
```

```python
#'Month','Year' 컬럼 자료형 바꾸기
df[['Month','Year']] = df[['Month','Year']].astype(int)
df.info()
```

```python
col_change = ['Coconut Oil Price Change','Olive Oil Price Change','Palm Kernel Oil Price Change','Palm Oil Price Change',
    'Peanut Oil Price Change', 'Rapeseed Oil Price Change' , 'Soybean Oil Price Change','Sunflower Oil Price Change']

#첫번째 행의 '-'들을 0으로 바꿔준다.
df[col_change] = df[col_change].replace('-',0)

# '%'를 *0.01 이라는 문자로 바꾸고, pd.eval함수를 적용해서 각 데이터 요소들을 문자형 상태에서 곱하기 연산을 해준다.
df[col_change] = df[col_change].replace('%','*0.01', regex=True).applymap(pd.eval)
```

```python
df.info()
```

```python
#select_dtypes 메서드를 사용해서 현재 object타입의 컬럼들만 price_col에 저장 (Change 컬럼들은 eval 함수를 적용했기때문에 float타입으로 이미 변경되었음을 알 수 있다.
price_col = df.select_dtypes(include=object).columns

#Price컬럼들의 ,를 제거하고 데이터타입을 float으로 변경 
df[price_col] = df[price_col].replace(',','', regex=True).astype(float)
```

```python
df
```

```python
#describe()메서드 사용하여 이상치 찾아보기
df[price_col].describe()
```

```python
# olive oil가격이 가장 많이 변화한 달을 .idxmax() 메서드를 사용하여 찾아보기
abs_max_idx = df["Olive Oil Price Change"].abs().idxmax()
df.loc[abs_max_idx-1:abs_max_idx]
```

2021년 1월에 olive oil 가격이 전달대비 +164% 변동되었다.

```python
#olive oil가격이 가장 적게 변한 달을 .idxmin() 메서드를 사용하여 찾아보기
#첫번째row은 변동령이 없으므로 제외
abs_min_idx = df["Olive Oil Price Change"][1:].abs().idxmin()
df.loc[abs_min_idx-1:abs_min_idx]
```

2011년 10월에 olive oil 가격이 가장 적게 변화함

```python
# olive oil가격이 가장 많이 떨어진 달을 .idxmin() 메서드를 사용하여 찾아보기
min_idx = df["Olive Oil Price Change"].idxmin()
df.loc[min_idx-1:min_idx]
```

2021년 11월에 olive oil 가격이 전달대비 -57% 변동되었다.

```python
#시각화 해보기

#마지막 row를 기준으로 컬럼을 내림차순 sort (legend 순서를 높은 가격순으로 맞춰주기 위해)
df_plt=df[price_col].sort_values(by=359, axis=1, ascending=False)

plt.figure(figsize=(20, 14))
plt.style.use('seaborn-whitegrid')
plt.plot(df_plt)
plt.legend(df_plt, fontsize=14)
plt.xticks(np.arange(5, 360, 12),labels=np.arange(1993, 2023, 1),fontsize=15,rotation=40)
plt.yticks(np.arange(500, 6001, 500),fontsize=17)
plt.xlabel("Year",fontsize=20)
plt.ylabel("Price (dollar per Ton)",fontsize=20)
plt.title("Historical Price Data of 8 Edible Oils",fontsize=30,pad=15)
plt.show()
```
