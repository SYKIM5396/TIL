```python
import pandas as pd

import numpy as np

from category_encoders import OneHotEncoder
from category_encoders import OrdinalEncoder
```

```python
df = pd.DataFrame({'ID':[1,2,3,4,5,6],
                'Color':['Red', 'Green', 'Blue', 'Blue', 'Green', 'Yellow']})
df
```

```python
ord_enc = OrdinalEncoder()
df_ord = ord_enc.fit_transform(df)
df_ord
```

```python
one_enc = OneHotEncoder(use_cat_names=True)
df_one = one_enc.fit_transform(df)
df_one
```
