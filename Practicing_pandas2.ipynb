```python
#isin 메서드 : 어떤 list에 존재하는 요소가 대상 DataFrame이나 Series에 존재 하는지를 True(존재), False(존재안함)로 반환
list1 =['Aug 1992','Dec 2000','Apr 2020']
df.loc[df.Month.isin(list1)]
```
